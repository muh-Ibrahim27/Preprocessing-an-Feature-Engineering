```python
# Handle missing values in 'age' by filling with the median
df['age'] = df['age'].fillna(df['age'].median())

# Handle missing values in categorical columns by filling with the mode
df['cabin'] = df['cabin'].fillna(df['cabin'].mode()[0])
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['profession'] = df['profession'].fillna(df['profession'].mode()[0])

# Display the number of remaining missing values to confirm
print("Missing values after imputation:")
print(df.isnull().sum())
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder() #Create Label Encoder

df['sex']=le.fit_transform(df['sex']) #Encode sex column
df['ticket_class']=le.fit_transform(df['ticket_class']) #Encode ticket_class column

print(df.head())
```

```python
#Categorical Feature Encoding(one-hot encoding)
#pd.get_dummies for one-hot encoding
encoded_df =pd.get_dummies(df['ticket_class'],prefix='ticket_class')

#Concatenate with original column and drop the old one
df=pd.concat([df,encoded_df],axis=1).drop(columns=['ticket_class'])

#Saved updated dataset
df.to_csv("cleaned_dataset.csv",index=False)

print("ticket_class to one-hot encoding and saved!")

```

```python
#Min-Max Scaling
from sklearn.preprocessing import MinMaxScaler

#Initialize the scalar
scaler=MinMaxScaler()

#Applying Min-Max scaling to 'age' and 'fare'
df_cleaned[['age','fare']]=scaler.fit_transform(df_cleaned[['age','fare']])

#Show scaled data
df_cleaned[['age','fare']].head()
```

```python
#Feature Engineering
# Extract title from 'Name' column
df['Title'] = df['name'].str.extract(r' ([A-Za-z]+)\.', expand=False)

# Replace rare titles with 'Rare'
rare_titles = ['Dr', 'Rev', 'Col', 'Major', 'Capt', 'Countess', 'Lady', 'Sir', 'Jonkheer', 'Don', 'Dona']
df['Title'] = df['Title'].replace(rare_titles, 'Rare')

# Combine similar titles
df['Title'] = df['Title'].replace(['Mlle', 'Ms'], 'Miss')
df['Title'] = df['Title'].replace('Mme', 'Mrs')

# FamilySize = SibSp + Parch + 1 (the +1 includes the passenger)
df['FamilySize'] = df['siblings_spouses'] + df['parents_children'] + 1

# Check results
print(df[['name', 'Title', 'siblings_spouses', 'parents_children', 'FamilySize']].head())
```
